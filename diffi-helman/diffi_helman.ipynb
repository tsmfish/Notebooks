–ú–∏ –±—É–¥–µ–º–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –≤–±—É–¥–æ–≤–∞–Ω—É –≤ Python —Ñ—É–Ω–∫—Ü—ñ—é `pow(base, exp, mod)`, —è–∫–∞ –¥—É–∂–µ –µ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –æ–±—á–∏—Å–ª—é—î $(base^{exp}) \pmod{mod}$.

```python
# –ö—Ä–æ–∫ 1: –ü—É–±–ª—ñ—á–Ω—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏ (–≤—ñ–¥–æ–º—ñ –≤—Å—ñ–º)
# p –º–∞—î –±—É—Ç–∏ –ø—Ä–æ—Å—Ç–∏–º —á–∏—Å–ª–æ–º
p = 23
# g –º–∞—î –±—É—Ç–∏ –ø–µ—Ä–≤—ñ—Å–Ω–∏–º –∫–æ—Ä–µ–Ω–µ–º –∑–∞ –º–æ–¥—É–ª–µ–º p
g = 5

print(f"–°–ø—ñ–ª—å–Ω—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏:\np = {p}\ng = {g}\n")

# –ö—Ä–æ–∫ 2: –ê–ª—ñ—Å–∞ –≥–µ–Ω–µ—Ä—É—î —Å–≤—ñ–π —Å–µ–∫—Ä–µ—Ç
# –¶–µ —á–∏—Å–ª–æ –ê–ª—ñ—Å–∞ —Ç—Ä–∏–º–∞—î –≤ —Ç–∞—î–º–Ω–∏—Ü—ñ!
a = 6
# –ê–ª—ñ—Å–∞ –æ–±—á–∏—Å–ª—é—î —Å–≤—ñ–π –ø—É–±–ª—ñ—á–Ω–∏–π –∫–ª—é—á A
A = pow(g, a, p)
print(f"–ê–ª—ñ—Å–∞ –æ–±—Ä–∞–ª–∞ —Å–µ–∫—Ä–µ—Ç a = {a}")
print(f"–ê–ª—ñ—Å–∞ –æ–±—á–∏—Å–ª—é—î A = (g^a) % p = ({g}^{a}) % {p} = {A}")
print(f"–ê–ª—ñ—Å–∞ –Ω–∞–¥—Å–∏–ª–∞—î –ë–æ–±—É A = {A}\n")

# –ö—Ä–æ–∫ 3: –ë–æ–± –≥–µ–Ω–µ—Ä—É—î —Å–≤—ñ–π —Å–µ–∫—Ä–µ—Ç
# –¶–µ —á–∏—Å–ª–æ –ë–æ–± —Ç—Ä–∏–º–∞—î –≤ —Ç–∞—î–º–Ω–∏—Ü—ñ!
b = 15
# –ë–æ–± –æ–±—á–∏—Å–ª—é—î —Å–≤—ñ–π –ø—É–±–ª—ñ—á–Ω–∏–π –∫–ª—é—á B
B = pow(g, b, p)
print(f"–ë–æ–± –æ–±—Ä–∞–≤ —Å–µ–∫—Ä–µ—Ç b = {b}")
print(f"–ë–æ–± –æ–±—á–∏—Å–ª—é—î B = (g^b) % p = ({g}^{b}) % {p} = {B}")
print(f"–ë–æ–± –Ω–∞–¥—Å–∏–ª–∞—î –ê–ª—ñ—Å—ñ B = {B}\n")

# --- –û–±–º—ñ–Ω –ø–æ "–Ω–µ–∑–∞—Ö–∏—â–µ–Ω–æ–º—É" –∫–∞–Ω–∞–ª—É ---
# –Ñ–≤–∞ –±–∞—á–∏—Ç—å p=23, g=5, A=8, B=19
# –ê–ª–µ –≤–æ–Ω–∞ –Ω–µ –º–æ–∂–µ –∑–Ω–∞–π—Ç–∏ 'a' –∞–±–æ 'b'

# –ö—Ä–æ–∫ 4: –ê–ª—ñ—Å–∞ —Ç–∞ –ë–æ–± –æ–±—á–∏—Å–ª—é—é—Ç—å —Å–ø—ñ–ª—å–Ω–∏–π —Å–µ–∫—Ä–µ—Ç

# –ê–ª—ñ—Å–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î –ø—É–±–ª—ñ—á–Ω–∏–π –∫–ª—é—á –ë–æ–±–∞ (B) —Ç–∞ —Å–≤—ñ–π —Å–µ–∫—Ä–µ—Ç (a)
S_Alice = pow(B, a, p)
print(f"–ê–ª—ñ—Å–∞ –æ–±—á–∏—Å–ª—é—î —Å–µ–∫—Ä–µ—Ç S = (B^a) % p = ({B}^{a}) % {p} = {S_Alice}")

# –ë–æ–± –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î –ø—É–±–ª—ñ—á–Ω–∏–π –∫–ª—é—á –ê–ª—ñ—Å–∏ (A) —Ç–∞ —Å–≤—ñ–π —Å–µ–∫—Ä–µ—Ç (b)
S_Bob = pow(A, b, p)
print(f"–ë–æ–± –æ–±—á–∏—Å–ª—é—î —Å–µ–∫—Ä–µ—Ç S = (A^b) % p = ({A}^{b}) % {p} = {S_Bob}")

# –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞
if S_Alice == S_Bob:
    print(f"\n‚úÖ –£—Å–ø—ñ—Ö! –°–ø—ñ–ª—å–Ω–∏–π —Å–µ–∫—Ä–µ—Ç–Ω–∏–π –∫–ª—é—á: {S_Alice}")
else:
    print(f"\n‚ùå –ü–æ–º–∏–ª–∫–∞! –ö–ª—é—á—ñ –Ω–µ –∑–±—ñ–≥–∞—é—Ç—å—Å—è.")
```

## ‚ö†Ô∏è –í–∞–∂–ª–∏–≤–µ –∑–∞—É–≤–∞–∂–µ–Ω–Ω—è: –ê—Ç–∞–∫–∞ "–õ—é–¥–∏–Ω–∞ –ø–æ—Å–µ—Ä–µ–¥–∏–Ω—ñ" (Man-in-the-Middle)

### üí° –û—Å–Ω–æ–≤–Ω–∞ —ñ–¥–µ—è:

### 1. –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è

### 2. –î–≤—ñ–π–∫–æ–≤–µ –ü—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è –ü–æ–∫–∞–∑–Ω–∏–∫–∞ –°—Ç–µ–ø–µ–Ω—è (b)

### 4. –ü–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å –û–±—á–∏—Å–ª–µ–Ω—å (–ü—ñ–¥–Ω–µ—Å–µ–Ω–Ω—è –¥–æ –ö–≤–∞–¥—Ä–∞—Ç—É)

### 5. –ê–ª–≥–æ—Ä–∏—Ç–º

3.  –ü–æ–∫–∏ $b > 0$:
    a.  **–ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –º–æ–ª–æ–¥—à–∏–π –±—ñ—Ç $b$**: `if (b % 2 == 1)` (–∞–±–æ `if (b & 1)`).
    b.  **–Ø–∫—â–æ –±—ñ—Ç == 1**: –ú–Ω–æ–∂–∏–º–æ –Ω–∞—à —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –ø–æ—Ç–æ—á–Ω—É –±–∞–∑—É: `result = (result * power) % n`.
    c.  **–ì–æ—Ç—É—î–º–æ –±–∞–∑—É –¥–ª—è –Ω–∞—Å—Ç—É–ø–Ω–æ–≥–æ –±—ñ—Ç–∞**: –ü—ñ–¥–Ω–æ—Å–∏–º–æ –±–∞–∑—É –¥–æ –∫–≤–∞–¥—Ä–∞—Ç—É: `power = (power * power) % n`.
    d.  **–ó—Å—É–≤–∞—î–º–æ –ø–æ–∫–∞–∑–Ω–∏–∫ $b$ –≤–ø—Ä–∞–≤–æ**: `b = b // 2` (–∞–±–æ `b >>= 1`), —â–æ–± –ø–µ—Ä–µ–π—Ç–∏ –¥–æ –Ω–∞—Å—Ç—É–ø–Ω–æ–≥–æ –±—ñ—Ç–∞.
4.  –ü–æ–≤–µ—Ä—Ç–∞—î–º–æ `result`.

## üßÆ –ú–∞—Ç–µ–º–∞—Ç–∏—á–Ω–∏–π –†–æ–∑–±—ñ—Ä –ù–∞–ø—Ä–∏–∫–ª–∞–¥: 7^25 mod 13
–ù–∞—à–µ –∑–∞–≤–¥–∞–Ω–Ω—è ‚Äî –æ–±—á–∏—Å–ª–∏—Ç–∏ $7^{25} \pmod{13}$.

### 1. –î–≤—ñ–π–∫–æ–≤–µ –ü—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è –°—Ç–µ–ø–µ–Ω—è

–°–ø–æ—á–∞—Ç–∫—É –ø—Ä–µ–¥—Å—Ç–∞–≤–∏–º–æ –ø–æ–∫–∞–∑–Ω–∏–∫ —Å—Ç–µ–ø–µ–Ω—è $b = 25$ —É –¥–≤—ñ–π–∫–æ–≤—ñ–π —Ñ–æ—Ä–º—ñ.
$$
25 = 16 + 8 + 1
$$
$$
25 = 1 \cdot 2^4 + 1 \cdot 2^3 + 0 \cdot 2^2 + 0 \cdot 2^1 + 1 \cdot 2^0
$$
–î–≤—ñ–π–∫–æ–≤–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è: **$11001_2$**

### 2. –†–æ–∑–∫–ª–∞–¥–∞–Ω–Ω—è –°—Ç–µ–ø–µ–Ω—è

–í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ —Ü–µ, –º–∏ –º–æ–∂–µ–º–æ —Ä–æ–∑–∫–ª–∞—Å—Ç–∏ $7^{25}$ –Ω–∞ –¥–æ–±—É—Ç–æ–∫ —á–ª–µ–Ω—ñ–≤:
$$
7^{25} = 7^{(16 + 8 + 1)} = 7^{16} \cdot 7^8 \cdot 7^1
$$
–¢–µ–ø–µ—Ä, –∑–≥—ñ–¥–Ω–æ –∑ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—è–º–∏ –º–æ–¥—É–ª—è—Ä–Ω–æ—ó –∞—Ä–∏—Ñ–º–µ—Ç–∏–∫–∏:
$$
7^{25} \pmod{13} \equiv (7^{16} \cdot 7^8 \cdot 7^1) \pmod{13}
$$

### 3. –ü–æ—Å–ª—ñ–¥–æ–≤–Ω–µ –ü—ñ–¥–Ω–µ—Å–µ–Ω–Ω—è –¥–æ –ö–≤–∞–¥—Ä–∞—Ç—É (mod 13)

–ù–∞–º –ø–æ—Ç—Ä—ñ–±–Ω–æ –∑–Ω–∞–π—Ç–∏ –∑–Ω–∞—á–µ–Ω–Ω—è $7^1$, $7^8$ —Ç–∞ $7^{16}$ –ø–æ –º–æ–¥—É–ª—é 13. –ú–∏ –º–æ–∂–µ–º–æ –µ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –∑–Ω–∞–π—Ç–∏ —ó—Ö, –ø–æ—Å–ª—ñ–¥–æ–≤–Ω–æ –ø—ñ–¥–Ω–æ—Å—è—á–∏ –æ—Å–Ω–æ–≤—É –¥–æ –∫–≤–∞–¥—Ä–∞—Ç—É, —ñ –≤—ñ–¥—Ä–∞–∑—É –±–µ—Ä—É—á–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–æ –º–æ–¥—É–ª—é 13 –Ω–∞ –∫–æ–∂–Ω–æ–º—É –∫—Ä–æ—Ü—ñ.

1.  **$7^1$**:
    $$7^1 \pmod{13} \equiv \mathbf{7}$$

2.  **$7^2$**:
    $$7^2 \pmod{13} \equiv (7^1)^2 \pmod{13} \equiv 7^2 \pmod{13} \equiv 49 \pmod{13} \equiv \mathbf{10}$$

3.  **$7^4$**:
    $$7^4 \pmod{13} \equiv (7^2)^2 \pmod{13} \equiv 10^2 \pmod{13} \equiv 100 \pmod{13} \equiv \mathbf{9}$$
    *(–û—Å–∫—ñ–ª—å–∫–∏ $13 \times 7 = 91$, $100 - 91 = 9$)*

4.  **$7^8$**:
    $$7^8 \pmod{13} \equiv (7^4)^2 \pmod{13} \equiv 9^2 \pmod{13} \equiv 81 \pmod{13} \equiv \mathbf{3}$$
    *(–û—Å–∫—ñ–ª—å–∫–∏ $13 \times 6 = 78$, $81 - 78 = 3$)*

5.  **$7^{16}$**:
    $$7^{16} \pmod{13} \equiv (7^8)^2 \pmod{13} \equiv 3^2 \pmod{13} \equiv 9 \pmod{13} \equiv \mathbf{9}$$

### 4. –§—ñ–Ω–∞–ª—å–Ω–µ –û–±—á–∏—Å–ª–µ–Ω–Ω—è

–¢–µ–ø–µ—Ä –º–∏ –±–µ—Ä–µ–º–æ –ª–∏—à–µ —Ç—ñ —á–ª–µ–Ω–∏, —è–∫—ñ –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—é—Ç—å –æ–¥–∏–Ω–∏—Ü—è–º —É –¥–≤—ñ–π–∫–æ–≤–æ–º—É –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—ñ **$11001_2$** ($2^{16}$, $2^8$, $2^1$):
* $7^1 \pmod{13} \equiv \mathbf{7}$
* $7^8 \pmod{13} \equiv \mathbf{3}$
* $7^{16} \pmod{13} \equiv \mathbf{9}$

–ü–µ—Ä–µ–º–Ω–æ–∂–∏–º–æ —ó—Ö –ø–æ –º–æ–¥—É–ª—é 13:
$$
7^{25} \pmod{13} \equiv (7^{16} \cdot 7^8 \cdot 7^1) \pmod{13}
$$
$$
\equiv (9 \cdot 3 \cdot 7) \pmod{13}
$$
$$
\equiv (27 \cdot 7) \pmod{13}
$$
–û—Å–∫—ñ–ª—å–∫–∏ $27 \pmod{13} \equiv 1$:
$$
\equiv (1 \cdot 7) \pmod{13}
$$
$$
\equiv \mathbf{7}
$$

**–í—ñ–¥–ø–æ–≤—ñ–¥—å:** $7^{25} \pmod{13} = 7$.

```python
## üõ†Ô∏è –†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –ê–ª–≥–æ—Ä–∏—Ç–º—É "–ü—ñ–¥–Ω–µ—Å–µ–Ω–Ω—è –¥–æ –°—Ç–µ–ø–µ–Ω—è —à–ª—è—Ö–æ–º –ü—ñ–¥–Ω–µ—Å–µ–Ω–Ω—è –¥–æ –ö–≤–∞–¥—Ä–∞—Ç—É"

def modular_pow(a: int, b: int, n: int) -> int:
    """
    –û–±—á–∏—Å–ª—é—î (a^b) % n –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –±—ñ–Ω–∞—Ä–Ω–æ–≥–æ –º–µ—Ç–æ–¥—É.

    a : –æ—Å–Ω–æ–≤–∞ (base)
    b : –ø–æ–∫–∞–∑–Ω–∏–∫ —Å—Ç–µ–ø–µ–Ω—è (exponent)
    n : –º–æ–¥—É–ª—å (modulus)
    """

    # –Ü–Ω—ñ—Ü—ñ–∞–ª—ñ–∑–∞—Ü—ñ—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
    result: int = 1

    # –Ü–Ω—ñ—Ü—ñ–∞–ª—ñ–∑–∞—Ü—ñ—è –±–∞–∑–∏ (a^1) % n
    # –ú–∏ –∑–≤–æ–¥–∏–º–æ 'a' –¥–æ –¥—ñ–∞–ø–∞–∑–æ–Ω—É [0, n-1]
    base: int = a % n

    print(f"–ü–æ—á–∏–Ω–∞—î–º–æ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è: ({a}^{b}) % {n}")
    print(f"–ü–æ—á–∞—Ç–∫–æ–≤–∏–π base = {base}, result = {result}\n")

    i: int = 0 # –õ—ñ—á–∏–ª—å–Ω–∏–∫ —ñ—Ç–µ—Ä–∞—Ü—ñ–π –¥–ª—è –≤—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—ó

    # –ì–æ–ª–æ–≤–Ω–∏–π —Ü–∏–∫–ª, –ø–æ–∫–∏ –ø–æ–∫–∞–∑–Ω–∏–∫ —Å—Ç–µ–ø–µ–Ω—è b –Ω–µ —Å—Ç–∞–Ω–µ 0
    while b > 0:
        # –û—Ç—Ä–∏–º—É—î–º–æ –¥–≤—ñ–π–∫–æ–≤–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è b (–¥–ª—è –ª–æ–≥—É–≤–∞–Ω–Ω—è)
        binary_b = bin(b)[2:]

        print(f"--- –Ü—Ç–µ—Ä–∞—Ü—ñ—è {i} ---")
        print(f"–ü–æ–∫–∞–∑–Ω–∏–∫ b = {b} (–±—ñ–Ω–∞—Ä–Ω–æ: {binary_b})")

        # –ö—Ä–æ–∫ 1: –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –º–æ–ª–æ–¥—à–∏–π –±—ñ—Ç b (b % 2)
        if (b % 2) == 1:
            # –Ø–∫—â–æ –±—ñ—Ç = 1, –º–Ω–æ–∂–∏–º–æ –Ω–∞—à —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –ø–æ—Ç–æ—á–Ω—É '–±–∞–∑—É'
            print(f"  -> –ú–æ–ª–æ–¥—à–∏–π –±—ñ—Ç = 1. –ú–Ω–æ–∂–∏–º–æ result:")
            print(f"     OLD result = {result}")
            print(f"     result = ({result} * {base}) % {n}")
            result = (result * base) % n
            print(f"     NEW result = {result}")
        else:
            print(f"  -> –ú–æ–ª–æ–¥—à–∏–π –±—ñ—Ç = 0. Result –Ω–µ –∑–º—ñ–Ω—é—î—Ç—å—Å—è ({result}).")

        # –ö—Ä–æ–∫ 2: –ü—ñ–¥–Ω–æ—Å–∏–º–æ '–±–∞–∑—É' –¥–æ –∫–≤–∞–¥—Ä–∞—Ç—É –¥–ª—è –Ω–∞—Å—Ç—É–ø–Ω–æ—ó —ñ—Ç–µ—Ä–∞—Ü—ñ—ó
        # –¶–µ (a^1 -> a^2), (a^2 -> a^4), (a^4 -> a^8) —ñ —Ç.–¥.
        print(f"  -> –û–Ω–æ–≤–ª—é—î–º–æ –±–∞–∑—É (–ø—ñ–¥–Ω–æ—Å–∏–º–æ –¥–æ –∫–≤–∞–¥—Ä–∞—Ç—É):")
        print(f"     OLD base = {base}")
        print(f"     base = ({base} * {base}) % {n}")
        base = (base * base) % n
        print(f"     NEW base = {base}")

        # –ö—Ä–æ–∫ 3: –ó—Å—É–≤–∞—î–º–æ –ø–æ–∫–∞–∑–Ω–∏–∫ b –≤–ø—Ä–∞–≤–æ (—Ü—ñ–ª–æ—á–∏—Å–µ–ª—å–Ω–µ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 2)
        # –¶–µ –µ–∫–≤—ñ–≤–∞–ª–µ–Ω—Ç–Ω–æ –≤–∏–¥–∞–ª–µ–Ω–Ω—é –º–æ–ª–æ–¥—à–æ–≥–æ –±—ñ—Ç–∞, —è–∫–∏–π –º–∏ –≤–∂–µ –æ–±—Ä–æ–±–∏–ª–∏
        b = b // 2

        print(f"  -> –ó—Å—É–≤–∞—î–º–æ –ø–æ–∫–∞–∑–Ω–∏–∫ b: new b = {b}\n")
        i += 1

    print(f"====================\n–¶–∏–∫–ª –∑–∞–≤–µ—Ä—à–µ–Ω–æ. b = 0.")
    print(f"–ö—ñ–Ω—Ü–µ–≤–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç: {result}")
    return result

print("\n–û–±—á–∏—Å–ª–µ–Ω–∏–π –Ω–∞–º–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç: %d" % modular_pow(7, 25, 13))
print("–ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ —á–µ—Ä–µ–∑ —Ñ—É–Ω–∫—Ü—ñ—é Python: %d" % pow(7, 25, mod=13))
```
